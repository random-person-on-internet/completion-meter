```python
import json
import os
import matplotlib.pyplot as plt

DATA_PATH = "../extracted_data_cleaned.json"
SAVE_DIR = "graphs_collector_style"
os.makedirs(SAVE_DIR, exist_ok=True)

def get_collector_curve(items):
    seen = set()
    curve = []
    count = 0
    for item in items:
        if item not in seen:
            seen.add(item)
            count+=1
        curve.append(count)
    return curve

with open(DATA_PATH, "r") as f:
    data = json.load(f)

for file_name, content in data.items():
    combined_items = content.get("entities", []) + content.get("noun_phrases", [])
    if not combined_items:
        continue
            
    curve = get_collector_curve(items)

    plt.figure(figsize=(6,3))
    plt.plot(curve, marker='o')
    plt.title(f"{file_name}")
    plt.xlabel("Position in List")
    plt.ylabel("Unique Count")
    plt.grid(True)
    plt.tight_layout()

    file_safe = file_name.replace("/","_").replace(".txt","")
    save_path = os.path.join(SAVE_DIR, f"{file_safe}_collector.png")
    plt.savefig(save_path)
    print(f"Graph for {file_name} saved")
    plt.close()
```
